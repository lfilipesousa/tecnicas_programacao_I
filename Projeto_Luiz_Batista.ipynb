```python
import pandas as pd
import numpy as np
from datetime import datetime, date
import statistics as sts
```

```python
# Criando os dataframes do investimento_1 e investimento_2

investimento_1 = pd.read_csv('investimento_parte1.csv', sep = ',')
investimento_2 = pd.read_csv('investimento_parte2.csv', sep = ',')

# Printando a quantidade de linhas de cada dataframe

print(f'A quantidade de linhas do dataframe investimento_1 é de: {investimento_1.shape[0]}, a quantidade de colunas é de: {investimento_1.shape[1]}')
print('-' * 32)
print(f'A quantidade de linhas do dataframe investimento_2 é de: {investimento_2.shape[0]}, a quantidade de colunas é de: {investimento_2.shape[1]}')
```

```python
# Informações do dataframe investimento_1
display(investimento_1)
print(investimento_1.nunique()) # Verificando se existem linhas repetidas
print('-' * 32)

# Informações do dataframe investimento_2
display(investimento_2)
print(investimento_2.nunique()) # Verificando se existem linhas repetidas
```

```python
# Juntando os dataframes investimento_1 e investimento_2 sem separar as linhas repetidas

df_total = pd.concat([investimento_1, investimento_2])
# Printando as linhas do dataframe

print(f' A quantidade de linhas dos dois dataframes juntos é de: {df_total.shape[0]}')

print('-' * 32)

# Utilizando o nunique para conferir o total de linhas únicas
print(f'A quantidade de linhas tirando as repetidas é de: {df_total.nunique()}')
df_total.head()
```

```python
# Juntando os dataframes investimento_1 e investimento_2 excluindo as linhas repetidas
df_investimento = df_total.sort_values('id')
df_investimento = df_investimento.drop_duplicates()

df_investimento.shape
```

```python
# Criando o dataframe para o arquivo cadastro
cadastro = pd.read_csv('cadastro.csv', sep = ',')
print(f'A quantidade de linhas do arquivo cadastro é de: {cadastro.shape[0]}')
```

```python
cadastro.head() # O arquivo cadastro tem algumas informações dos usuários
```

```python
# Criando um dataframe juntando o arquivo cadastro com o arquivo investimento
df_invcad = pd.merge(cadastro, df_investimento, how = 'outer', on = 'id')

# Ordenando o dataframe pela coluna id de forma crescente
df_invcad = df_invcad.sort_values('id')

# Printando as 5 primeiras linhas 
df_invcad.head()

```

```python
# Criando função para porcentagem de NaNs
def percent_nan(df):
    pn = (df.isnull().sum()*100/len(df)).sort_values(ascending = False)
    return pn
```

```python
# Percentuais de NaNs no dataframe por columns
percent_nan(df_invcad)

```

```python
# Para tratar esses dados NaNs, irei substituir os dados da coluna ativo por '0', que simboliza que não foi investido nenhuma renda nesse determinado ativo.
# Ativos_x

df_invcad.iloc[:, 11:18] = df_invcad.iloc[:, 11:18].fillna(0)
df_invcad.iloc[:, 11:18].head()

# Verificando se a substituição foi feita corretamente

df_invcad.iloc[:, 11:18].isna().sum()
```

```python
# Nas colunas de nota, irei utilizar a média para a substituição.
# Notas_x
df_invcad[['Nota1', 'Nota2']] = df_invcad[['Nota1', 'Nota2']].fillna(0)

# Nota1_x, criando a média e fazendo a substituição

media_1 = (df_invcad['Nota1'].mean())
df_invcad.loc[df_invcad.Nota1.isnull(), 'Nota1'] = media_1

# Nota2_x, criando a média e fazendo a substituição

media_2 = (df_invcad['Nota2'].mean())
df_invcad.loc[df_invcad.Nota2.isnull(), 'Nota2'] = media_2

# Verificando a soma dos números iguais a zero, para ver se a substituição foi feita corretamente

df_invcad[['Nota1', 'Nota2']].isnull().sum()
```

```python
# Nos perfis irei utilizar moda, por se tratar de strings de classificação.
# Perfil_x

print(df_invcad['Perfil'].describe(include=['o']))
print('-' * 32)

# O dado com maior frequência é o 'Moderado', então irei substituir os NaNs por ele.

df_invcad['Perfil'] = df_invcad['Perfil'].fillna(0)
df_invcad.loc[df_invcad.Perfil.isnull(), 'Perfil'] = 'Agressivo'

# Verificando os valores null

df_invcad['Perfil'].isnull().sum()
```

```python
# Retirando os cifrões da coluna salário_x
df_invcad['Salario'] = df_invcad['Salario'].str.replace('$', '')

# Preenchendo os NaN com zeros
df_invcad['Salario'] = df_invcad['Salario'].fillna(0)
df_invcad['Salario'].tail()
```

```python
# Deixando a coluna salário em modo numérico.
# Não irei substituir os dados nulos, não serão usados para a análise e existem muitos dados faltantes.

df_invcad['Salario'] = df_invcad['Salario'].astype(float)

# Renomeando a coluna Salário para Renda

df_invicad = df_invcad.rename(columns={'Salario': 'Renda'}, inplace = True)

print(df_invcad['Renda'])
```

```python
# Criando variável splitando as strings da coluna email

dominio = df_invcad['Email'].str.split('@', n = 1, expand = True)

# Criando a coluna 'Dominio' e selecionando o index [1] que corresponde aos dominios de todos os emails

df_invcad['Dominio'] = dominio[1]
df_invcad['Dominio']
```

```python
# Criar coluna média das notas1 e notas2

df_invcad['Media'] = (df_invcad['Nota1'] + df_invcad['Nota2']) / 2
df_invcad['Media']
```

```python
# Criar coluna total de todos os ativos

# Ativos

# Separei todos os ativos em uma variável para simplificar na hora da adição da coluna

ativos = np.array(df_invcad.iloc[:, 11:18].astype(float))
ativos = np.sum(ativos, axis = 1)
ativos = pd.DataFrame(ativos)
ativos
# Adicionando a coluna Total_ativos_x

df_invcad['Total_ativos'] = ativos
df_invcad['Total_ativos']
```

```python
# Transformando a coluna de Média para inteiro

df_invcad['Media'] =  np.round(df_invcad['Media'].astype(int))
```

```python
# Mudando os dados da coluna 'Sexo' para Masculino e Feminino

df_invcad.loc[df_invcad.Sexo == 'Fem', 'Sexo'] = 'Feminino'
df_invcad.loc[df_invcad.Sexo == 'Masc', 'Sexo'] = 'Masculino'
df_invcad['Sexo']
```

```python
# Criando coluna 'NPS' e classificando de acordo com a média das notas.

df_invcad.loc[df_invcad['Media'] >= 80, 'NPS'] = 'Promotor'
df_invcad.loc[df_invcad['Media'] < 50, 'NPS'] = 'Detrator'
df_invcad.loc[df_invcad['NPS'].isna(), 'NPS'] = 'Neutro'

df_invcad['NPS'].value_counts()
```

```python
# Criando função para deixar a idade pela data de nascimento atualizada.

df_invcad['Idade'].head()
def age(born):
    born = datetime.strptime(born, '%Y-%m-%d').date()
    today = date.today()
    return today.year - born.year - ((today.month, 
                                      today.day) < (born.month, 
                                                    born.day))

# Aplicando a função na column 'Idade'.

df_invcad['Idade'] = df_invcad['Nasc'].apply(age)  

# printando a idade do investidor mais velho

idade = df_invcad['Idade'].sort_values()
print(f'O investidor mais velho tem: {idade[0]} anos')
```

```python
# Calculando Mediana da coluna Média
mediana = sts.median(df_invcad['Media'])
print(f'A mediana da coluna média é: {mediana}')

```
